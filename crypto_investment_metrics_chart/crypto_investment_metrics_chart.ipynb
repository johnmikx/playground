```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for the Chart
cryptos = ['Cardano (ADA)', 'Avalanche(AVAX)', 'Chainlink (LINK)']
market_caps = [28.8, 12, 11] # in $B
roi_potential = [75, 115, 90] # average of ROI range (%)
risk_scores = [3.1, 3.2, 2.8] # from risk assessment

# Set Up the Plot
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.25
index = np.arange(len(cryptos))

# Plot Bars
bars1 = plt.bar(index, market_caps, bar_width,
                label='Market Cap ($B)', color='skyblue')
bars2 = plt.bar(index + bar_width, roi_potential, bar_width,
                label='1-Year ROI Potential (%)', color='lightgreen')
bars3 = plt.bar(index + 2 * bar_width, risk_scores, bar_width,
                label='Risk Score', color='salmon')

# Customize the Chart
plt.xlabel('Cryptocurrencies', fontweight='bold')
plt.title('Comparison of Cryptocurrency Investment Metrics (July 2025)', fontweight='bold')
plt.xticks(index + bar_width, cryptos)
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Add Value Labels on Top of Bars
for bars in [bars1, bars2, bars3]:
  for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height}',
             ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
